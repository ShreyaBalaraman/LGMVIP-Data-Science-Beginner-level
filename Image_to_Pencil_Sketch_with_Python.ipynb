Importing Libraries

```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

Reading the image

```python
dog_image = cv2.imread("dog_image.jpg")
cv2_imshow(dog_image)
```

Read the image in RBG format

```python
rbg = cv2.cvtColor(dog_image, cv2.COLOR_BGR2RGB)
cv2_imshow(rbg)
```

convert RBG format to grayscale image

```python
gray_scale = cv2.cvtColor(rbg, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_scale)
```

Inverted grayscale image(Negative image)

```python
inverted_gray_scale = 255 - gray_scale
cv2_imshow(inverted_gray_scale )
```

Blurry image

```python
blur_image = cv2.GaussianBlur(inverted_gray_scale, (21,21), 0)
cv2_imshow(blur_image)
```

Inverted blurry image

```python
inverted_blur = 255 - blur_image
cv2_imshow(inverted_blur)
```

Pencil Sketch

```python
pencil_sketch = cv2.divide(gray_scale, inverted_blur, scale=255)
cv2_imshow(pencil_sketch)
```
